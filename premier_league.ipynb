```python
import numpy as np

# Normalize numerical columns
cols_to_normalize = ['gf', 'ga', 'xg', 'xga', 'poss', 'sh', 'sot']
for col in cols_to_normalize:
    liverpool_2025[col + '_norm'] = (liverpool_2025[col] - liverpool_2025[col].min()) / (liverpool_2025[col].max() - liverpool_2025[col].min())

# Venue adjustment factor
liverpool_2025['venue_factor'] = np.where(liverpool_2025['venue'].str.lower() == 'home', 1.05, 1.0)

# Performance metric
liverpool_2025['performance_metric'] = liverpool_2025['venue_factor'] * (
    0.30*liverpool_2025['gf_norm']
    - 0.20*liverpool_2025['ga_norm']
    + 0.20*liverpool_2025['xg_norm']
    - 0.10*liverpool_2025['xga_norm']
    + 0.10*liverpool_2025['poss_norm']
    + 0.05*liverpool_2025['sh_norm']
    + 0.05*liverpool_2025['sot_norm']
)
```

```python
liverpool_2025[['performance_metric', 'venue']] \
    .round(2) \
    .sort_values(by='performance_metric', ascending=False)
```

```python
big_game_performance = liverpool_2025.loc[['Arsenal', 'Manchester City', 'Manchester Utd', 'Tottenham', 'Chelsea', 'Newcastle Utd'],['performance_metric', 'venue']].round(2).sort_values(by='performance_metric', ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(big_game_performance.index, big_game_performance['performance_metric'], marker='o')
plt.xticks(rotation=45)  # Rotate for readability
plt.xlabel('Opponent')
plt.ylabel('Performance Metric')
plt.title('Big Game Performance')
plt.tight_layout()
plt.show()
```

```python
# 1. Ensure opponent is a column
if 'opponent' not in big_game_performance.columns:
    big_game_performance = big_game_performance.reset_index()
```

```python
# 2. Pivot so each opponent has Home & Away as separate columns
pivot_df = big_game_performance.pivot_table(
    index='opponent',
    columns='venue',
    values='performance_metric',
    aggfunc='first'  # or 'mean' if multiple rows exist
)
```

```python

# 3. Compute total for sorting
pivot_df['total'] = pivot_df.sum(axis=1)
```

```python
# 4. Sort by total descending
pivot_df = pivot_df.sort_values(by='total', ascending=False)

```

```python
# 5. Prepare for plotting
opponents = pivot_df.index
x = np.arange(len(opponents))
bar_width = 0.4

```

```python
# 6. Plot bars (align properly now)
plt.figure(figsize=(10, 5))
plt.bar(x - bar_width/2, pivot_df['Home'], width=bar_width, color='blue', label='Home')
plt.bar(x + bar_width/2, pivot_df['Away'], width=bar_width, color='red', label='Away')
plt.xticks(x, opponents, rotation=45)
plt.xlabel('Opponent')
plt.ylabel('Performance Metric')
plt.title('Big Game Performance (Home vs Away)')
plt.legend()
plt.tight_layout()
plt.savefig("big_game_performance.png")
plt.show()

```
